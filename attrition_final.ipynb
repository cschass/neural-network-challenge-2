## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
X_df = attrition_df.drop(columns=['Attrition', 'Department'])
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = ['Age', 'DistanceFromHome', 'EnvironmentSatisfaction', 'JobSatisfaction', 'NumCompaniesWorked', 'RelationshipSatisfaction', 'TotalWorkingYears', 'YearsAtCompany',
                    'YearsInCurrentRole', 'YearsWithCurrManager']

# Create X_df using your selected columns
X_df = attrition_df[selected_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
missings = X_train.isnull().sum()
print(missings)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python

X_train.dtypes
```

```python
from re import X
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Scale the training and testing data
print("Scaled Training Data: ", X_train_scaled[:5])
print("Testing Data: ", X_test[:5])
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
department_trained = department_encoder.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_department_train = department_encoder.transform(y_train[['Department']])
y_department_test = department_encoder.transform(y_test[['Department']])
```

```python
print("Y data:", y_department_train[:5])
print("Y data test: ", y_department_test[:5])
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
attrition_train = attrition_encoder.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_attrition_train = attrition_encoder.transform(y_train[['Attrition']])
y_attrition_test = attrition_encoder.transform(y_test[['Attrition']])
```

```python
print("Y data:", y_attrition_train[:5])
print("Y data test: ", y_attrition_test[:5])
```

## Part 2: Create, Compile, and Train the Model

```python

# Find the number of columns in the X training data.
input_features = X_train.shape[1]
print(input_features)

# Create the input layer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
model = Sequential()
model.add(Dense(units=10, activation='relu'))

# Create at least two shared layers
model.add(Dense(units=8, activation='softmax'))
model.add(Dense(units=5, activation='softmax'))
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_input = Input(shape=(input_features,))

# Create the hidden layer
department_hidden = Dense(units=10, activation='relu')(department_input)

# Create the output layer
department_output = Dense(units=3, activation='softmax')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_input = Input(shape=(input_features,))
# Create the hidden layer
attrition_hidden = Dense(units=10, activation='relu')(attrition_input)

# Create the output layer
attrition_output = Dense(units=2, activation='softmax')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=[department_input, attrition_input], outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=[['accuracy'], ['accuracy']])

# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit([X_train_scaled, X_train_scaled], [y_department_train, y_attrition_train], epochs=100, batch_size=32)
```

```python
# Evaluate the model with the testing data
eval_results = model.evaluate([X_test_scaled, X_test_scaled], [y_department_test, y_attrition_test])
```

```python
# Print the accuracy for both department and attrition
print("Department Accuracy:", eval_results[0])
print("Attrition Accuracy:", eval_results[1])
```
